```python
for i in range(len(blueshell)):
    # Definir encabezado
    encabezado = ['Producto','Calibre','Valor','Unidad de medida','Definición','Archivo']

    blueshell[i].columns = encabezado

    #Limpiar la tabla
    listado = [None,
     'Carne',
     'Media Concha',
     'Entero',
     'Rendimiento Olla mmpp Propia',
     'Rendimiento Olla mmpp Terceros']

    blueshell[i].reset_index(drop=True, inplace=True)

    blueshell[i].Producto = blueshell[i].Producto.apply(lambda p: p if p in listado else 'Eliminar')
    blueshell[i].drop(blueshell[i].loc[blueshell[i].Producto == 'Eliminar'].index, axis=0, inplace=True)

    blueshell[i].Producto.fillna(method='ffill', inplace = True)

    listado_2 = [
     'Carne',
     'Media Concha',
     'Entero',
     'Rendimiento Olla mmpp Propia',
     'Rendimiento Olla mmpp Terceros']

    blueshell[i].Producto = blueshell[i].Producto.apply(lambda p: p if p in listado_2 else 'Eliminar')
    blueshell[i].drop(blueshell[i].loc[blueshell[i].Producto == 'Eliminar'].index, axis=0, inplace=True)

    blueshell[i]['Mes'] = i+1
    blueshell[i]['Año'] = 2021
    blueshell[i]['Empresa'] = 'Blue Shell'
    blueshell[i].drop(['Archivo'], axis=1, inplace= True)
    blueshell[i].drop(blueshell[i].loc[blueshell[i]['Definición'].isnull()].index, axis=0, inplace=True)

    blueshell_out = blueshell_out.append(blueshell[i])
```
