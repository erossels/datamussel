```python
#Lectura de datos
chonchi = pd.DataFrame()
empresa = 'chonchi'

for f in folders:
    try:
        os.chdir(kpi+f)
    except:
        print('Error al ingresar a', f)
        continue
    files = os.listdir(kpi+f)
    for file in files:
        if empresa in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(kpi+f+'/'+file, sheet_name='1. Ingreso mmpp')
                        
                        #Anexar
                        chonchi = chonchi.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde est√° la tabla
                        wb = load_workbook(filename = kpi+f+'/'+file)
                        hoja = '1. Ingreso mmpp'
                        sheet = wb[hoja]
                        print(sheet)

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values)
                        df['ARCHIVO'] = file

                        #Anexar
                        chonchi = chonchi.append(df)
                    except:
                        print('Error en', file)
                    
```
