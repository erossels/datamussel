### CAMANCHACA

```python
#Camanchaca
camanchaca = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'camanchaca' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    print('skip')
                    #p.save_book_as(file_name=prices+str(year)+'/'+str(w)+'/'+file ,dest_file_name=prices+str(year)+'/'+str(w)+'/'+file[:-4]+'.xlsx')        
                    #file = prices+str(year)+'/'+str(w)+'/'+file[:-4]+'.xlsx'
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde está la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file)
                        hoja = wb.sheetnames
                        sheet = wb[hoja[0]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 

                        #Limitar el df
                        df.reset_index(inplace = True, drop=True)

                        #Anexar
                        camanchaca = camanchaca.append(df)
                    except:
                        print('Error en', file)
                    

# Resetear indices
```

```python
camanchaca.sample(10)
```

```python
# Corregir problemas de lectura
camanchaca[camanchaca[1].isnull()] = camanchaca[camanchaca[1].isnull()].shift(-1, axis=1)
camanchaca[camanchaca[1].apply(lambda x: str(x).find('w') != -1)] = camanchaca[camanchaca[1].apply(lambda x: str(x).find('w') != -1)].shift(-1, axis=1)
```

```python
# Nombrar columnas
camanchaca.drop([0,10,11,12,13,14], axis=1, inplace=True)
camanchaca.columns = ['FECHA','PRODUCTO','EMPAQUE','PAIS','CALIBRE','UNIDAD PESO','MONEDA','CANTIDAD','PRECIO CFR']
camanchaca.dropna(inplace=True, axis=0)
```

```python
camanchaca.sample(20)
```

```python
camanchaca.loc[camanchaca.FECHA == '2021-07-01']
```

```python
# Transformaciones
camanchaca.drop(labels=list(camanchaca[camanchaca.FECHA.isnull()].index), axis=0, inplace=True)
camanchaca.FECHA = camanchaca.FECHA.apply(lambda f: pd.to_datetime(f,errors='coerce'))
camanchaca.CANTIDAD = camanchaca.CANTIDAD.apply(lambda c: pd.to_numeric(c,errors='coerce'))
camanchaca['PRECIO CFR'] = camanchaca['PRECIO CFR'].apply(lambda p: pd.to_numeric(p,errors='coerce'))
camanchaca.CALIBRE = camanchaca.CALIBRE.apply(lambda x: x if pd.isnull(x) else x.split(" ")[0])
camanchaca.PAIS = camanchaca.PAIS.apply(str.capitalize)
camanchaca.reset_index(drop = True, inplace=True)
```

```python
camanchaca.drop(camanchaca[camanchaca.FECHA.isna()].index, axis=0, inplace=True)
```

```python
# Eliminar duplicados
camanchaca.drop_duplicates(inplace=True)
# Creación de campos
camanchaca['EMPRESA'] = 'Camanchaca'
camanchaca['SEMANA'] = camanchaca.FECHA.apply(lambda d: d.isocalendar()[1])
```

```python
# Chequeo
camanchaca.info()
camanchaca.sample(20)
```

### St. Andrews

```python
companies
```

```python
#Lectura de datos
standrews = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'st andrews' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(prices+str(year)+'/'+str(w)+'/'+file, skiprows=1)
                        #Anexar
                        standrews = standrews.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde está la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file)
                        hoja = wb.sheetnames
                        sheet = wb[hoja[0]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 
                        df.dropna(axis=0, thresh=2, inplace=True)

                        #Anexar
                        standrews = standrews.append(df)
                    except:
                        print('Error en', file)
                    
```

```python
standrews.head()
```

```python
# Completar valores
standrews['FECHA'].fillna(standrews['Fecha Carga Estimada'], inplace = True)
standrews['PRODUCTO'].fillna(standrews['Producto Global'], inplace = True)
standrews['Empaque'].fillna(standrews['EMPAQUE'], inplace=True)
standrews['CALIBRE'].fillna(standrews['Calibre'], inplace = True)
standrews['CANTIDAD'].fillna(standrews['Peso'], inplace = True)
standrews['PRECIO CFR'].fillna(standrews['Precio Unitario'], inplace = True)
standrews['DESTINO'].fillna(standrews['Pais'], inplace = True)
```

```python
standrews.drop(['Fecha Carga Estimada','Producto Global','EMPAQUE','Calibre','Peso','Precio Unitario','Pais'], axis=1, inplace = True)
```

```python
standrews.dropna(axis=0, thresh=7, inplace=True)
```

```python
standrews['TIPO VENTA'].fillna(standrews['VENTA'], inplace = True)
standrews['TIPO VENTA'].fillna(standrews['VENTA '], inplace=True)
standrews['TIPO VENTA'].fillna(standrews['VENTAS'], inplace=True)
```

```python
standrews['TIPO VENTA'].isna().sum()
```

```python
standrews.drop(['VENTA', 'VENTA ', 'VENTAS'], axis=1, inplace=True)
```

```python
#Lectura de datos
sudmaris = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'sudmaris' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(prices+str(year)+'/'+str(w)+'/'+file, skiprows=1)
                        
                        #Anexar
                        sudmaris = sudmaris.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde está la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file)
                        hoja = wb.sheetnames
                        sheet = wb[hoja[1]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 

                        #Anexar
                        sudmaris = sudmaris.append(df)
                    except:
                        print('Error en', file)
                    
```

```python
sudmaris.head(10)
```

```python
# Eliminar fila semanas y Nan
sudmaris.drop(sudmaris.iloc[:,[0,10]], axis=1, inplace=True)

#Corregir nombres de columnas
sudmaris.columns = sudmaris.iloc[0,:]
sudmaris.reset_index(inplace=True, drop=True)
sudmaris.drop(sudmaris.iloc[0].index.name,inplace= True)
```

```python
# Modificaciones columnas 
sudmaris.columns = ['FECHA','PRODUCTO','EMPAQUE','CALIBRE','CANTIDAD','PRECIO CFR', 'PRECIO FOB','TIPO VENTA', 'PAIS']
sudmaris['EMPRESA'] = 'Sudmaris'
sudmaris.FECHA = sudmaris.FECHA.apply(lambda f: pd.to_datetime(f, errors='coerce'))
sudmaris.drop(sudmaris.loc[sudmaris.FECHA.isna()].index, axis=0, inplace = True)
```

```python
# Transformaciones en columnas
sudmaris['SEMANA'] = sudmaris.FECHA.apply(lambda f: f.isocalendar()[1])
sudmaris.CALIBRE = sudmaris.CALIBRE.apply(lambda x: str(x).split(" ")[0])
sudmaris.CANTIDAD = sudmaris.CANTIDAD.apply(lambda c: pd.to_numeric(c, errors='coerce'))
sudmaris['PRECIO CFR'] = sudmaris['PRECIO CFR'].apply(lambda c: pd.to_numeric(c, errors='coerce'))
sudmaris['PRECIO FOB'] = sudmaris['PRECIO FOB'].apply(lambda c: pd.to_numeric(c, errors='coerce'))
sudmaris.sort_values(by='FECHA', ascending = True, inplace = True)
sudmaris.info()
```

### LANDES

```python
#Lectura de datos
landes = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'landes' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(prices+str(year)+'/'+str(w)+'/'+file, skiprows=1)
                        
                        #Anexar
                        landes = landes.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde está la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file)
                        hoja = wb.sheetnames
                        print(hoja)
                        sheet = wb[hoja[1]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 

                        #Anexar
                        landes = sudmaris.append(df)
                    except:
                        print('Error en', file)
                    
```

```python
landes.head()
```

```python
# Reordenar datos
landes.drop(landes.iloc[:,[0]], axis=1, inplace=True)
```
