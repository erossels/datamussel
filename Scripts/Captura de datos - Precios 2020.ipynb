### CAMANCHACA

```python
#Camanchaca
camanchaca = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'camanchaca' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(prices+str(year)+'/'+str(w)+'/'+file)
                        #Anexar
                        camanchaca = camanchaca.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde est√° la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file)
                        hoja = wb.sheetnames
                        sheet = wb[hoja[0]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 

                        #Limitar el df
                        df.reset_index(inplace = True, drop=True)

                        #Anexar
                        camanchaca = camanchaca.append(df)
                    except:
                        print('Error en', file)
                    

```

### St. Andrews

```python
#Lectura de datos
standrews = pd.DataFrame()

for w in weeks:
    try:
        os.chdir(prices+str(year)+'/'+str(w))
    except:
        print('Error al ingresar a', w)
        continue
    w_files = os.listdir(prices+str(year)+'/'+str(w)+'/')
    for file in w_files:
        if 'st andrews' in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(prices+str(year)+'/'+str(w)+'/'+file, skiprows=1)
                        #Anexar
                        df['ARCHIVO'] = file
                        standrews = standrews.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde est√° la tabla
                        wb = load_workbook(filename = prices+str(year)+'/'+str(w)+'/'+file, skiprows=1)
                        hoja = wb.sheetnames
                        sheet = wb[hoja[0]]

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values) 
                        df.dropna(axis=0, thresh=2, inplace=True)
                        df['ARCHIVO'] = file

                        #Anexar
                        standrews = standrews.append(df)
                    except:
                        print('Error en', file)
                    
```

```python
# Completar valores
standrews['EMPAQUE'].fillna(standrews['Empaque'], inplace=True)
standrews['VENTA'].fillna(standrews['Venta'], inplace=True)
standrews['Precio CFR'].fillna(standrews['Precio Unitario'], inplace = True)
standrews['Precio CFR'].fillna(standrews['Precio Unitario CFR/kilo'], inplace = True)
standrews['FOB por Kilo'].fillna(standrews['FOB x Kilo'], inplace = True)
standrews['FOB por Kilo'].fillna(standrews['Unnamed: 8'], inplace = True)
standrews['VENTA'].fillna(standrews['VENTA '], inplace = True)
standrews['VENTA'].fillna(standrews['VENTAS'], inplace = True)
standrews['Pais'].fillna(standrews['Mercado'], inplace = True)
```

```python
standrews.drop(['Empaque','Venta','Precio Unitario','Precio Unitario CFR/kilo','Mercado', 'FOB x Kilo','Venta FOB','Unnamed: 8','VENTAS','VENTA ','Precio CFR'], axis=1, inplace=True)
standrews
```
