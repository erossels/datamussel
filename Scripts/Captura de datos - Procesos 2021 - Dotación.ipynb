## DOTACION

```python
#Lectura de datos
chonchi = pd.DataFrame()
empresa = 'chonchi'

for f in folders:
    try:
        os.chdir(kpi+f)
    except:
        print('Error al ingresar a', f)
        continue
    files = os.listdir(kpi+f)
    for file in files:
        if empresa in file.lower():
            print(file)
            if file[-4:] != '.lnk':
                # Casos XLS
                if file[-4:] in ['.xls','.slk'] :
                    try:
                        df = pd.read_excel(kpi+f+'/'+file, sheet_name='3. Dotación')
                        
                        #Anexar
                        df['ARCHIVO'] = file
                        chonchi = chonchi.append(df)
                        
                    except:
                        print('Error en', file)
                else: # Casos XLSX
                    try: 
                        #Leer hoja donde está la tabla
                        wb = load_workbook(filename = kpi+f+'/'+file)
                        hoja = '3. Dotación'
                        sheet = wb[hoja]
                        print(sheet)

                        #Transformar hoja en un df
                        df = pd.DataFrame(sheet.values)
                        df['ARCHIVO'] = file

                        #Anexar
                        chonchi = chonchi.append(df)
                    except:
                        print('Error en', file)
                    
```

### Sudmaris

```python
#Lectura de datos
sudmaris = pd.DataFrame()
empresa = 'sudmaris'

for f in folders:
    try:
        os.chdir(kpi+f)
    except:
        print('Error al ingresar a', f)
        continue
    files = os.listdir(kpi+f)
    for file in files:
        if empresa in file.lower():
            try:
                print(file)
                df = pd.read_excel(kpi+f+'/'+file, sheet_name='3. Dotación')
                df['ARCHIVO'] = file
                #Anexar
                sudmaris = sudmaris.append(df)

            except:
                print('Error en', file)

```

```python
sudmaris_enero = sudmaris.iloc[:,[0,1,2]]
sudmaris_febrero = sudmaris.iloc[:,[0,3,4]]
sudmaris_marzo = sudmaris.iloc[:,[0,5,6]]
sudmaris_abril = sudmaris.iloc[:,[0,7,8]]
sudmaris_mayo = sudmaris.iloc[:,[0,9,10]]
sudmaris_junio = sudmaris.iloc[:,[0,11,12]]
sudmaris_julio = sudmaris.iloc[:,[0,13,14]]
sudmaris_agosto = sudmaris.iloc[:,[0,15,16]]
sudmaris_septiembre = sudmaris.iloc[:,[0,17,18]]
sudmaris_octubre = sudmaris.iloc[:,[0,19,20]]
sudmaris_noviembre = sudmaris.iloc[:,[0,21,22]]
sudmaris_diciembre = sudmaris.iloc[:,[0,23,24]]

sudmaris = [sudmaris_enero, sudmaris_febrero, sudmaris_marzo, sudmaris_abril, sudmaris_mayo, sudmaris_junio,\
           sudmaris_julio, sudmaris_agosto, sudmaris_septiembre, sudmaris_octubre, sudmaris_noviembre,\
           sudmaris_diciembre]

sudmaris_out = pd.DataFrame()
```

```python
encabezado_sud = ['Cargo','Contrato Plazo Fijo','Contrato Plazo Indefinido']
mes = 1

for sud in sudmaris:
    #Depurar cargos
    
    sud.columns = encabezado_sud
    sud.Cargo = sud.Cargo.apply(lambda c: c if c in listado_cargos else 'Eliminar')
    sud.drop(sud.loc[sud.Cargo == 'Eliminar'].index, axis=0, inplace=True)

    # Completar nones con 0
    sud.fillna(0, axis=1, inplace=True)
    sud.reset_index(inplace = True, drop=True)

    #Ultimas columnas
    sud['Año'] = 2021
    sud['Empresa'] = 'Sudmaris'
    sud['Mes'] = mes
    
    mes +=1
    
    sudmaris_out = sudmaris_out.append(sud)


sudmaris_out.drop_duplicates(inplace=True)
sudmaris_out.reset_index(inplace = True, drop=True)
sudmaris = sudmaris_out.copy()
```
